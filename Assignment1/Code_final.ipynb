```python
df_11 = df_1.drop(columns = ['APN', 'PROJECT NUMBER', 'NAME', 'SITE #','DATE FUNDED','DEVELOPER','MANAGEMENT COMPANY','CONTACT PHONE','PHOTO','JOBS','PROJECT SUMMARY URL','CONTRACT NUMBERS','SITE LONGITUDE','SITE LATITUDE','GPS_COORDS ON MAP','SITE COMMUNITY','DATE STAMP'])
```

```python
import re

# Extract ZIP code (assuming 5-digit format) from SITE ADDRESS
df_11['ZIP CODE'] = df_11['SITE ADDRESS'].apply(lambda x: re.findall(r'\b\d{5}\b', x)[-1] if re.search(r'\b\d{5}\b', x) else None)

# Display the extracted ZIP codes
print(df_11['ZIP CODE'].head())
```

```python
df_11 = df_11.drop(columns = ['SITE ADDRESS'])
```

```python
df_11['ZIP CODE'] = df_11['ZIP CODE'].astype('int')
df_11.info()
```

```python
# Grouping data by 'ZIP CODE' and summing the values
aggregated_data = df_11.groupby('ZIP CODE').sum()

# Generate numerical indices for the x-axis
x_indices = range(len(aggregated_data))

# Now plot the aggregated data using numerical indices instead of ZIP codes
plt.figure(figsize=(10, 6))
plt.bar(x_indices, aggregated_data['SITE UNITS'], label='SITE UNITS')
plt.bar(x_indices, aggregated_data['SH UNITS PER SITE'], label='SH UNITS PER SITE', bottom=aggregated_data['SITE UNITS'])

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Total Units')
plt.title('Bar Chart of Total Units by ZIP CODE (Numerical Index)')
plt.legend()

# Display the plot
plt.show()

```
