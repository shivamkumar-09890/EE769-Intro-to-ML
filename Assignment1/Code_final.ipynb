## Check for unexpected values
[NavBar](#NavBar)

```python
#sorce-->{3}: NLP lecture by standford. 1st lecture of this serie has all understanding of regular expression which was used in bellow code.

import re

# Define a function to check for unexpected values
def check_unexpected_values(df, column, pattern=None):
    # Convert column to string if pattern is provided
    if pattern:
        df[column] = df[column].astype(str)
        # Check for values that do not match the expected pattern
        unexpected_values = df[~df[column].apply(lambda x: bool(pattern.match(x)))]
    else:
        # Convert to string and check for empty strings or NaN values
        df[column] = df[column].astype(str)
        unexpected_values = df[df[column].str.strip().eq('') | df[column].isna()]
    
    return unexpected_values

# Define patterns for columns where a specific format is expected
patterns = {
    'facility_name': re.compile(r'^[\w\s\W]+$'),  # Letters, digits, spaces, and special characters
    'score': re.compile(r'^\d+(\.\d+)?$'),  # Numeric values (integers or decimals)
    'grade': re.compile(r'^[A-F]$'),  # Letter grades (A-F)
    'service_code': re.compile(r'^[A-Z0-9]+$'),  # Alphanumeric codes
    'service_description': None,  # No specific pattern, just check for empty strings and NaNs
    'employee_id': re.compile(r'^EE\d{7}$'),  # Numeric employee IDs
    'pe_description': None,  # No specific pattern, just check for empty strings and NaNs
    'program_element_pe': None,  # No specific pattern, just check for empty strings and NaNs
    'program_name': None,  # No specific pattern, just check for empty strings and NaNs
    'program_status': re.compile(r'^[A-Za-z\s]+$'),  # Only letters and spaces
    'ZIP CODE': re.compile(r'^\d{5}(-\d{4})?$')  # Valid ZIP or ZIP+4 format
}

# Check unexpected values for each column
for column in df_21.columns:  # Iterate over each column in the DataFrame df_21.
    pattern = patterns.get(column)  # Retrieve the pattern associated with the current column from the patterns dictionary.
    unexpected_values = check_unexpected_values(df_21, column, pattern)  # Check for unexpected values in the column using the specified pattern.
    
    if not unexpected_values.empty:  # If there are unexpected values in the column (i.e., the result is not empty).
        print(f"Unexpected values in column '{column}':")  # Print a message indicating there are unexpected values in the column.
        print(unexpected_values)  # Print the unexpected values.
        print("\n")  # Print a newline for better readability between columns.
    else:  # If there are no unexpected values in the column.
        print(f"No unexpected values in column '{column}'.")  # Print a message indicating there are no unexpected values in the column.
        print("\n")  # Print a newline for better readability between columns.


```
